## Importing the libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Load and Prepare Data

```python
data=pd.read_csv('googleplaystore.csv')
data.shape
data.columns
data.head()
data.info()
data.describe()
```

## EDA

```python
data.isnull().sum()
```

## Data Preprocessing

```python
num_cols=data.select_dtypes(include=['float64','int64']).columns
for col in num_cols:
  data[col].fillna(data[col].mean(),inplace=True)


cat_cols=data.select_dtypes(include=['object']).columns
for col in cat_cols:
  data[col].fillna(data[col].mode()[0] ,inplace=True)

  data.isnull().sum()
```

## Strorytelling - Visualization

```python
sns.histplot(data['Rating'], bins=20, kde=True)
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

sns.countplot(data='App',x=)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='Category', data=data, order=data['Category'].value_counts().index)
plt.title("Number of Apps per Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Reviews',y='Rating',data=data,alpha=0.5)
plt.title(" Ratings vs Reviews")
plt.xlabel("Nummber of Reviews")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Size',y='Rating',data=data,alpha=0.5)
plt.title(" App size vs Reviews")
plt.xlabel("Nuber of Reviews")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = data.corr(numeric_only=True)   #   
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
